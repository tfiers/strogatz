```python
def udot(u, r, h):
    return - u**3 / 3 + r*u + h
    
def plot(ax, r, h, ulim=5, ylim=(-20, 20)):
    ax.axhline(0, c='black')
    U = np.linspace(-ulim, ulim, 1000)
    ax.plot(U, udot(U, r, h), label=f'r = {r:g},  h = {h:g}')
    ax.set_xlabel(r'$u$')
    ax.set_ylabel(r'$\frac{du}{dt}$')
    ax.set_ylim(ylim)

fig, ax = plt.subplots(figsize=(6,6))
plot(ax, 0., -0.3)
ax.legend();
```
